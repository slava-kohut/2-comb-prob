## Problem 1

```python
from math import factorial
def choose(n,k):
    return factorial(n)/(factorial(k)*factorial(n-k))
```

```python
choose(7,3)
```

```python
from itertools import combinations
for i in combinations(range(1,8),3):
    print(i)
```
